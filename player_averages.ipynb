```python
import pandas as pd
import numpy as np
from nba_api.stats.endpoints.leaguedashplayerstats import LeagueDashPlayerStats
from constants import *
from point_values import point_values
```

```python
player_stats_object = LeagueDashPlayerStats(league_id_nullable = league_id, season = current_season, season_type_all_star = season_type)
player_stats_df = player_stats_object.get_data_frames()[0]

# reducing dataframe to necessary information
player_stats_df = player_stats_df[player_stats_cols]

fg_miss = player_stats_df['FGA'] - player_stats_df['FGM'] # calculate field goals missed
fg3_miss = player_stats_df['FG3A'] - player_stats_df['FG3M'] # calculate three pointers missed
ft_miss = player_stats_df['FTA'] - player_stats_df['FTM'] # calculate free throws missed

# insert new columns in appropriate location in dataframe
player_stats_df.insert(loc = 8, column = 'FGMS', value = fg_miss)
player_stats_df.insert(loc = 11, column = 'FG3MS', value = fg3_miss)
player_stats_df.insert(loc = 14, column = 'FTMS', value = ft_miss)

# calculate fantasy points
player_stats_df.head()
```

```python
# calculate fantasy points
player_stats_df['FP'] = (
    player_stats_df['FGM'] * point_values['FGM']['point_value'] +
    player_stats_df['FGMS'] * point_values['FGMS']['point_value'] +
    player_stats_df['FTM'] * point_values['FTM']['point_value'] +
    player_stats_df['FTMS'] * point_values['FTMS']['point_value'] +
    player_stats_df['FG3M'] * point_values['FG3M']['point_value'] +
    player_stats_df['REB'] * point_values['REB']['point_value'] +
    player_stats_df['AST'] * point_values['AST']['point_value'] +
    player_stats_df['STL'] * point_values['STL']['point_value'] +
    player_stats_df['BLK'] * point_values['BLK']['point_value'] +
    player_stats_df['TOV'] * point_values['TOV']['point_value'] +
    player_stats_df['PTS']
)
player_stats_df.head()
```
